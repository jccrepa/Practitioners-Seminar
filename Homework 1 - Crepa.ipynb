```python
import pandas as pd
import matplotlib.pyplot as plt

prices = pd.read_csv("prices.csv", index_col=False)
prices.head()
```

```python
prices_filtered = prices.loc[(prices.Ticker == 'NVDA') & (prices.Country == 'usa') &
         (prices.Date >= "2000-01-03") & (prices.Date <= "2005-12-31")]

prices_filtered
```

```python
prices_filtered = prices_filtered.sort_values("Date", ascending=True).reset_index(drop=True)
prices_filtered["Time"] = range(1, len(prices_filtered) + 1)

prices_filtered
```

```python
prices_plot = prices_filtered.set_index("Time")

prices_filtered[["Open", "High", "Low", "Close"]].plot(figsize=(12,6))

plt.xlabel("Time")
plt.ylabel("Price")
plt.title("NVDA Stock Prices (2000–2005)")
plt.grid(True)
plt.show()
```

```python
prices_filtered[["Volume"]].plot(figsize=(12,6))

plt.xlabel("Time")
plt.ylabel("Volume")
plt.title("NVDA Volume (2000–2005)")
plt.grid(True)
plt.show()
```
